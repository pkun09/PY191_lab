```python
# finding a value of pi by the area of a circle and a square -- Monto-Carlo simulation
import random
n = 0

for i in range (10000000):
    x = -1 + 2 * random.random()
    y = -1 + 2 * random.random()

    if x ** 2 + y **2 <= 1:
        n += 1

Value = (4*n)/10000000
print (Value)
```

```python
sentence = "It was the best of times, it was the worst of times... in short, the period was so far like the present period, that some of its noisiest authorities insisted on its being received, for good or for evil, in the superlative degree of comparison only."
words = sentence.split()

for i, word in enumerate(words): 
    if i % 2 == 1:
        words[i] = word[::-1]

print(" ".join(words))

## try again to make every other word backwards in the sentence
## include the enumerate function, use the index
```

```python
import numpy as np
temperatures_f = np.array([58.2, 64.0, 69.3, 74.1, 77.8, 76.2, 71.0, 65.4])
print("Mean:", temperatures_f.mean())
print("Standard deviation:", temperatures_f.std())

temps = temperatures_f - temperatures_f.mean()
temps = np.abs(temps)

print("Mean Absolute Deviation around Mean:", np.sum(temps)/ len(temps))

median = np.quantile(temperatures_f, 0.50)
new_temps = temperatures_f - median
new_temps = np.abs(new_temps)
print("Mean Absolute Deviation around Median:", np.sum(new_temps)/ len(new_temps))

third = np.quantile(temperatures_f, 0.75)
first = np.quantile(temperatures_f, 0.25)
print("Q2 is:", (third - first)/2)

```

```python
temp_2d = np.array([[64.0, 66.5, 70.2], [77.1, 75.6, 71.3]])
print(temp_2d.mean(axis=0)) ##axis specifies which direction you want to calculate the mean
print(temp_2d.mean(axis=0, keepdims=True))
print(temp_2d.mean(axis=0).shape)
print(temp_2d.mean(axis=0, keepdims=True).shape)

## now we want minimum along the columns
print(temp_2d.min(axis=0))
```

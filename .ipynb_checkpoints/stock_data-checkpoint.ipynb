```python
# -*- coding: utf-8 -*-
### LLY
"""
Corrected code to download stock data using yfinance.
"""

import os
import pandas as pd
import yfinance as yf

def get_stock(ticker, start_date, end_date, s_window, l_window):
    """Retrieve stock data and compute indicators."""
    try:
        # Download stock data
        df = yf.download(ticker, start=start_date, end=end_date)
        if df.empty:
            print(f"No data found for {ticker}")
            return None

        # Calculate daily returns
        df['Return'] = df['Close'].pct_change().fillna(0)

        # Process date attributes
        df.reset_index(inplace=True)
        df['Date'] = pd.to_datetime(df['Date'])
        df['Year'] = df['Date'].dt.year
        df['Month'] = df['Date'].dt.month
        df['Day'] = df['Date'].dt.day
        df['Weekday'] = df['Date'].dt.day_name()
        df['Week_Number'] = df['Date'].dt.strftime('%U')
        df['Year_Week'] = df['Date'].dt.strftime('%Y-%U')

        # Round price columns
        price_cols = ['Open', 'High', 'Low', 'Close']
        df[price_cols] = df[price_cols].round(2)

        # Calculate moving averages
        df['Short_MA'] = df['Close'].rolling(window=s_window, min_periods=1).mean()
        df['Long_MA'] = df['Close'].rolling(window=l_window, min_periods=1).mean()

        # Reorder columns
        col_order = [
            'Date', 'Year', 'Month', 'Day', 'Weekday', 'Week_Number', 'Year_Week',
            'Open', 'High', 'Low', 'Close', 'Volume', 'Return',
            'Short_MA', 'Long_MA'
        ]
        return df[col_order]

    except Exception as e:
        print(f"Error processing {ticker}: {str(e)}")
        return None

# Example usage
if __name__ == "__main__":
    try:
        ticker = "LLY"
        output_file = os.path.join(os.getcwd(), f"{ticker}.csv")

        # Get stock data
        df = get_stock(
            ticker=ticker,
            start_date="2021-06-30",
            end_date="2026-06-30",
            s_window=14,
            l_window=50
        )

        # Save to CSV if data exists
        if df is not None:
            df.to_csv(output_file, index=False)
            print(f"Successfully saved {len(df)} records to {output_file}")
        else:
            print("No data available to save.")

    except Exception as e:
        print(f"Main execution error: {str(e)}")

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
stocks = pd.read_csv("LLY.csv")
stocks = stocks.drop(stocks.index[0])
##stocks.head()

fig, ax1 = plt.subplots(figsize=(3, 3))

stocks["Close"] = stocks["Close"].astype(float)


ax1.plot(stocks["Date"], stocks["Close"], color='#1f77b4', linestyle='-', linewidth=1.0, label='Close (Date)' )


plt.show()

## i dont know how to fix this

## next exercise

monday = stocks[stocks["Weekday"] == "Monday"]
tuesday = stocks[stocks["Weekday"] == "Tuesday"]
wednesday = stocks[stocks["Weekday"] == "Wednesday"]
thursday = stocks[stocks["Weekday"] == "Thursday"]
friday = stocks[stocks["Weekday"] == "Friday"]

monday_mean = monday['Return'].mean()
print("Monday:", monday_mean)

tuesday_mean = tuesday['Return'].mean()
print("Tuesday:", tuesday_mean)

wednesday_mean = wednesday['Return'].mean()
print("Wednesday:", wednesday_mean)

thursday_mean = thursday['Return'].mean()
print("Thursday:", thursday_mean)

friday_mean = friday['Return'].mean()
print("Friday:", friday_mean)

stocks = pd.read_csv('LLY.csv')
close = 'Close'

# 2. Extract the last digit of the float: 
# abs() handles any negative floats 
# np.floor ensures strict mathematical flooring
last_digits = np.floor(abs(df['Close']) * 10).astype(int) % 10

# 3. Count the frequency of each number
counts = last_digits.value_counts().sort_index()

print(counts)

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
stocks = pd.read_csv("LLY.csv")
stocks = stocks.drop(stocks.index[0])

st_ret = stocks["Return"]
st_ret.head()

pos = 0
neg = 0

for returns in st_ret:
    if returns > 0:
        pos += 1
    if returns < 0:
        neg += 1

print(pos)
print(neg)

mean = np.mean(st_ret)
std_dev = np.std(st_ret)

lower_bound = mean - (2 * std_dev)
upper_bound = mean + (2 * std_dev)

outliers = 0

for returns in st_ret:
    if returns < lower_bound:
        outliers +=1
    if returns > upper_bound:
        outliers +=1 
print(outliers, "/", len(st_ret))

```
